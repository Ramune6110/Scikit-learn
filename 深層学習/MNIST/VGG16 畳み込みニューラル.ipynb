```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

```python
# 0～9の手書き文字のデータセット(MNIST)
mnist = keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
# 画像データの正規化
train_images_norm, test_images_norm = train_images / 255.0, test_images / 255.0
model = tf.keras.models.Sequential([
    tf.keras.layers.Reshape((28, 28, 1), input_shape=(28, 28)),  
    tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu',name="Conv2D_1"), 
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),name="MaxPooling2D_1"),    
    tf.keras.layers.Flatten(),                         
    tf.keras.layers.Dense(10, activation='softmax')   
])
# 最適化手法, 損失関数の評価基準, 精度表示
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
# 学習
result = model.fit(train_images_norm, train_labels, epochs=5,batch_size=32)
```

```python
#TensorBoard
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir="./log",histogram_freq=1)
result = model.fit(train_images_norm, train_labels, epochs=5,validation_data=(test_images_norm, test_labels),
                   batch_size=32,callbacks=[tensorboard_callback]) 
#VGG16
model_vgg16 = tf.keras.applications.vgg16.VGG16(weights='imagenet')
model_vgg16.summary()
```

```python
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from pprint import pprint

# 画像を読み込む
img_dog  = load_img('チワワ.jpg', target_size=(224, 224))
img_cat  = load_img('ネコ.jpg', target_size=(224, 224))
img_bird = load_img('ノスリ.jpg', target_size=(224, 224))
plt.rcParams['figure.figsize']=(5,5)
plt.rcParams['figure.dpi']=75
# 画像を表示
fig = plt.Figure(figsize=(5,5),dpi=75)
plt.imshow(img_dog)
plt.show()
fig = plt.Figure(figsize=(5,5),dpi=75)
plt.imshow(img_cat)
plt.show()
fig = plt.Figure(figsize=(5,5),dpi=75)
plt.imshow(img_bird)
plt.show()
# 画像をベクトルに変換
img_dog  = img_to_array(img_dog)
img_cat  = img_to_array(img_cat)
img_bird = img_to_array(img_bird)
# vgg16モデルに画像を入力
img_dog  = tf.keras.applications.vgg16.preprocess_input(img_dog)
img_cat  = tf.keras.applications.vgg16.preprocess_input(img_cat)
img_bird = tf.keras.applications.vgg16.preprocess_input(img_bird)
# vgg16のモデルを使用して入力画像の予測と結果の表示
predict = model_vgg16.predict(img_dog[tf.newaxis, ...])
result  = tf.keras.applications.vgg16.decode_predictions(predict, top=5)
pprint(result)
predict = model_vgg16.predict(img_cat[tf.newaxis, ...])
result  = tf.keras.applications.vgg16.decode_predictions(predict, top=5)
pprint(result)
predict = model_vgg16.predict(img_bird[tf.newaxis, ...])
result  = tf.keras.applications.vgg16.decode_predictions(predict, top=5)
pprint(result)
```

```python
#転移学習
import tensorflow
from tensorflow.keras.optimizers import SGD
baseModel = tf.keras.applications.vgg16.VGG16(weights='imagenet', include_top=False, 
                                              input_tensor=tensorflow.keras.Input(shape=(224, 224, 3)))

model2 = tf.keras.Sequential([
    baseModel,
    tf.keras.layers.Flatten(), 
    Dense(128, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

for layer in baseModel.layers:
    layer.trainable = False

model2.compile(optimizer=SGD(lr=0.0001, momentum=0.9), loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow.keras.preprocessing.image as Image

datagen = Image.ImageDataGenerator(
            featurewise_center            = False,
            samplewise_center             = False,
            featurewise_std_normalization = False,
            samplewise_std_normalization  = False,
            zca_whitening                 = False,
            rotation_range                = 30,
            width_shift_range             = 0.3,
            height_shift_range            = 0.3,
            horizontal_flip               = True,
            vertical_flip                 = False,
            validation_split              = 0.1
        )

train_datagen = Image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.vgg16.preprocess_input
)

validation_datagen = Image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.vgg16.preprocess_input
)

BATCH_SIZE = 1

train_generator = train_datagen.flow_from_directory(
   '17flowers/train',
   target_size=(224, 224),
   batch_size=BATCH_SIZE,
   class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
   '17flowers/validation',
   target_size=(224, 224),
   batch_size=BATCH_SIZE,
   class_mode='categorical'
)

model2.compile(optimizer=SGD(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
result = model2.fit(
    train_generator,
    steps_per_epoch=train_generator.n // BATCH_SIZE,
    validation_data=validation_generator,
    validation_steps=validation_generator.n // BATCH_SIZE,
    epochs=6
)
```

```python
import matplotlib.pyplot as plt
epochs = 6

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(range(1, epochs + 1), result.history['accuracy'], label="accuracy",c='r')
ax2.plot(range(1, epochs + 1), result.history['loss'], label="loss",c='b')
ax1.set_ylabel('Accuracy')
ax2.set_ylabel('Loss')
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()

ax1.legend(handler1 + handler2, label1 + label2, loc=2, borderaxespad=0.)

plt.show()
```
